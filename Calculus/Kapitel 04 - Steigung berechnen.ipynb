## 2. Konstante Steigung (Wiederholung) <a class="anchor" id="2"/>

In Kapitel 2 hatten wir die **konstante Steigung** im Abschnitt über lineare Funktionen (Kapitel 2.3) kennengelernt. Die Steigung gibt an, um wieviele Einheiten sich die y-Koordinate eines Punktes verändert, wenn sich die x-Koordinate dieses Punktes um eine Einheit verändert. "Konstante Steigung" bedeutet, dass die Steigung für jeden Punkt einer Funktion <u>gleich</u> ist (Abbildung 1). *Lineare Funktionen* oder *konstante Funktionen* besitzen eine konstante Steigung. Kurvenförmige Funktionen wie *quadratische Funktionen* oder *Exponentialfunktionen* besitzen eine **nicht konstante Steigung**, welche an verschiedenen Punkten des Funktionsgraphen <u>verschieden</u> steil ist (Abbildung 2).<br><br>

Für die Berechnung der **konstanten Steigung** brauchen wir den sogenannten **Differenzenquotienten**. Die Formel dafür lautet:<br>

$ m = \dfrac{\Delta y}{\Delta x} = \dfrac{y_2-y_1}{x_2 - x_1} $<br>

${\Delta y}$ und ${\Delta x}$ sind Kurzschreibweisen des Bruches $\frac{y_2-y_1}{x_2 - x_1} $ und beschreiben die <b>Differenz</b> zwischen den jeweiligen Koordinaten der beiden ausgewählten Punkte. Diese Differenzen können als Linien im Graph eingezeichnet werden, wodurch sich das sogenannte **Steigungsdreieck** ergibt:<br>
<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/c/c1/Wiki_slope_in_2d.svg/800px-Wiki_slope_in_2d.svg.png" style="width: 200px;"/>

Um die konstante Steigung zu berechnen, nehmen wir zwei beliebige Punkte $P(x_1,y_1)$ und $Q(x_2,y_2)$, die auf der Geraden der jeweiligen Funktion liegen und setzen diese in die **Differenzenquotienten**-Formel ein. Im folgenden Graphen wären das die Punkte $P(0,2)$ und $Q(1, 4)$. 

<img src="img/function_types/linear_function_2_two_labeled_points.png" align="center"  style="width: 300px; height: 300px;" /><br>


Durch das Einsetzen in die Formel erhalten wir die Steigung $m = 2$.

$ m=\dfrac{4-2}{1-0}=\dfrac{2}{1}=2$<br>

Hat man eine Formel einer linearen oder konstanten Funktion vorliegen, ist das "Berechnen" der Steigung noch einfacher, da der Wert lediglich abgelesen werden muss. Bei der linearen Funktion ist dies der Wert vor dem $x$. Bei der Gleichung $f(x) = 4x + 7$ wäre die Steigung $m=4$. Eine konstante Funktion steigt oder fällt nicht, die Steigung ist immer $m=0$.[<sup>1</sup>](#fn1)

<hr style="border: 0.1px solid black;"/>
<span id="fn1" style="font-size:8pt; line-height:1"><sup style="font-size:5pt">1</sup>&nbsp; <u>Beweis/Herleitung</u>: Als lineare Funktion ausgedrückt würde die konstante Funktion $f(x) = 5$ folgendermaßen aussehen: $f(x) = 0x + 5$, woran man wieder die Steigung ablesen könnte.</span><br>

## 3. Nicht konstante Steigung <a class="anchor" id="3"/>

Funktionen, deren Graphen **Kurven** sind, besitzen eine **nicht konstante Steigung**, was bedeutet, dass die Steigung in verschiedenen Punkten des Graphen verschieden steil ist. Schauen wir uns dafür einmal die Abbildung 4 an. Wir stellen uns dabei vor, dass wir auf einem Fahrrad fahren und die Exponentialfunktion ein Berg ist. Zu Beginn ist der Berg noch nicht so steil und wir müssen uns nicht wirklich anstrengen. Umso weiter wir aber den Berg hochfahren, umso steiler wird er und umso schwieriger wird die Fahrt. Bei der linearen Funktion aus Abbildung 3 ist die Fahrt immer gleich schwer, da die Steigung immer gleich steil ist. In Abbildung 5 wechselt sogar das Vorzeichen der Steigung, je nachdem, wo wir uns auf dem Graphen befinden. Befinden wir uns links von der y-Achse, **fällt** der Graph (= Vorzeichen der Steigung ist *negativ*), befinden wir uns jedoch rechts von der y-Achse, **steigt** der Graph (= Vorzeichen der Steigung ist *positiv*).

Wir halten also fest, dass die Steigung bei Kurven an verschiedenen Punkten unterschiedlich steil ist. Unser Ziel ist es, eine Methode bzw. Funktion zu finden, mithilfe derer wir die Steigung <u>eines</u> Punktes einer Kurve ermitteln können. Die Berechnung der Steigung mithilfe des **Differenzenquotienten** aus dem vorherigen Abschnitt können wir nun eigentlich nicht mehr anwenden, um die Steigung der Kurve in einem Punkt zu ermitteln, da wir im vorherigen Abschnitt gesehen hatten, dass man dafür zwei Punkte benötigt. Wir werden jedoch sehen, dass wir die Differenzenquotienten-Formel modifizieren können, um nur die Steigung von einem Punkt zu erhalten. Zunächst schauen wir uns aber an, wie wir die Differenzenquotienten-Formel auf Funktionen anwenden können, deren Graphen Kurven sind.

### Die Berechnung der Steigung mithilfe von zwei Punkten

Für die Berechnung der Steigung mithilfe von zwei Punkten betrachten wir die quadratische Funktion $f(x) = x^2 + 1$. Wie auch bei der konstanten Steigung bestimmten wir zwei Punkte und zeichnen diese in den Graphen ein: $P(0,1)$ und $Q(1,2)$ (siehe Abbildung 6). Diese verbinden wir mit einer **Sekante**[<sup>2</sup>](#fn2) (Abbildung 7). Die Sekante ist eine lineare Funktion, welche eine konstante Steigung besitzt. Diese können wir mithilfe des Steigungsdreieck (Abbildung 8) herleiten, indem wir die Punkte in die Formel für den **Differenzenquotienten** einsetzen: 

$\dfrac{\Delta y}{\Delta x} = \dfrac{y_2-y_1}{x_2 - x_1} = \dfrac{2 - 1}{1 - 0} = 1$


Die Steigung ist also $m = 1$, die Formel für die Sekante in den Abbildungen ist demnach $f(x) = x + 1$.

### Modifikation der Differenzenquotienten-Formel

Mithilfe des Differenzenquotienten können wir zwar die Steigung von zwei Punkten berechnen, wir wollen aber die Steigung <u>eines</u> Kurvenpunkts berechnen. Dazu benötigen wir den **Differentialquotienten**, der auf dem Differenzenquotienten aufbaut. Die Idee des Differentialquotienten ist, dass wir, um die Steigung von einem Punkt $P$ berechnen zu können, die Distanz zwischen dem Punkt $P$ und einem weiteren Punkt $Q$ verringern und so die Punkte **annähern**. In Abbildung 9 ist zu erkennen, dass durch die Annäherung die Sekante zu einer **Tangenten**[<sup>3</sup>](#fn3) wird. Durch die Annäherung wird der Abstand der beiden Punkte immer geringer, bis er schließlich gegen $0$ tendiert, was bedeutet, dass er unglaublich klein ist und bei Berechnungen wie eine $0$ behandelt wird. Diese Annäherung ist mathematisch durch den **Grenzwert** definiert. Sie wird als **Differentialquotient** bezeichnet. Die Formel[<sup>4</sup>](#fn4) dafür ist eine erweiterte Form der Differenzenquotienten-Formel, da der Differentialquotient der Grenzwert des Differenzenquotienten ist:

$ m = \lim \limits_{x_1 \to x_0} \dfrac{y_1 - y_0}{x_1 - x_0} $ &nbsp; oder alternativ &nbsp; $ m = \lim \limits_{x_1 \to x_0} \dfrac{f(x_1) - f(x_0)}{x_1 - x_0} $

Mithilfe des Differentialquotienten ist es uns möglich, dass wir die Steigung eines Punktes einer Kurve berechnen können, da wir wie beim Differenzenquotienten eine lineare Funktion ermitteln, von welcher wir einfach die Steigung berechnen können. Nun ist das Problem bei nicht konstanten Steigungen, dass die Steigungen der Punkte an verschiedenen Stellen der Kurve unterschiedlich sind. Theoretisch müsste man für jeden Punkt die Steigung mithilfe des Differentialquotienten berechnen. Da dies unglaublich zeitaufwendig wäre, gibt es die **Ableitungsfunktion**, die im folgenden Abschnitt erläutert wird. Dort werden wir auch sehen, wie wir mit der Differentialquotienten-Formel rechnen können.

## 4. Ableitungsfunktion und h-Methode <a class="anchor" id="4"/>

Die **Ableitungsfunktion** oder einfach nur **Ableitung** $f'(x)$ einer Funktion $f(x)$ ist eine Funktion, die jedem $x$-Wert der Funktion $f(x)$ ihren Differentialquotienten zuordnet. Somit muss für die Ermittlung der Steigung eines Kurvenpunktes dieser Punkt nur in die Ableitungsfunktion eingesetzt werden, eine zusätzliche Berechnung mit der Differentialquotienten-Formel entfällt. Diesen Ableitungsfunktionen und ihren Rechenregeln werden wir uns in den nächsten Kapiteln widmen. In diesem Abschnitt soll zunächst untersucht werden, wie man von der bereits bekannten Differentialquotienten-Formel zu der Ableitungsfunktion gelangt. Dafür wird die sogenannte **h-Methode** verwendet. Diese hat weniger einen praktischen Nutzen, sondern dient eher dem Verständnis der Herkunft der Ableitungsfunktionen. Dieses Verständnis ist aber wichtig, da die Ableitungen das Herzstück dieser Tutorialreihe sind und es soll mit der h-Methode gezeigt werden, dass Ableitungen keine abstrakten Rechenwerkzeuge sind, sondern eigentlich nur eine konformere Art und Weise, die Steigungen von Punkten eines kurvenförmigen Graphen zu berechnen.

Die **h-Methode** erweitert die Differenzenquotienten-Formel:

$ m = \dfrac{f(x_1) - f(x_0)}{x_1 - x_0} $ &nbsp;&nbsp; &rarr; &nbsp;&nbsp;  $ \dfrac{f(x + h) - f(x)}{h} $


Die Idee der h-Methode ist, dass man keinen speziellen Punkt $x_1$ in den Differenzenquotienten einsetzt, sondern einen Platzhalter $h$ (daher auch der Name *h*-Methode). Dieser Platzhalter steht für einen Abstand zweier Werte und sieht wie folgt aus:

$ h = x_1 - x_0 $ 


Dies geben wir in die Differenzenquotienten-Formel ein:

$ m = \dfrac{f(x_1) - f(x_0)}{x_1 - x_0} $ &nbsp;&nbsp; &rarr; &nbsp;&nbsp;  $ \dfrac{f(x_0 + h) - f(x_0)}{h} $ &nbsp;&nbsp; &rarr; &nbsp;&nbsp;  $ \dfrac{f(x + h) - f(x)}{h} $<br>


$f(x_1)$ wird hier zu $f(x_0 + h)$. Dafür formen wir die Gleichung $h = x_1 - x_0$ zu $ x_1 = x_0 + h$ um und setzen diese in $f(x_1)$ ein. Den Nenner ersetzen wir durch $h$. Da nun kein $x_1$ mehr in der Formel vorkommt, können wir anstatt $x_0$ auch $x$ schreiben. Wie das Steigungsdreieck für die Sekante mit der h-Methode aussieht, wird durch Abbildung 10 deutlich.

Da der Differenzenquotient Teil des Differentialquotienten ist, können wir die h-Methode auch für den Differentialquotient verwenden. Die Formel sieht wie folgt aus:

$ m = \lim \limits_{h \to 0} \dfrac{f(x + h) - f(x)}{h} $

$h$ ist hier ein Wert, der gegen $0$ tendiert. Dadurch nähern sich $x$ und $x+h$ an, je kleiner $h$ ist. $h$ ist dabei nie $0$, da nicht durch $0$ geteilt werden darf. Trotzdem ignorieren wir nach der Berechnung der Ableitungsfunktion $h$, da es so nahe an $0$ liegt, dass es für die weiteren Berechnungen kaum relevant ist. $h$ ist für uns also nur ein temporäres Werkzeug. In Abbildung 11 wird der Vorgang des immer kleiner werdenden $h$'s demonstriert. Die anfängliche Sekante wird nach und nach zur Tangente. Unsere neue Formel erlaubt es uns nun, eine Funktion zu berechnen, die jedem $x$ den Wert ihres Differentialquotienten zuordnet. Genau das ist das Ziel einer **Ableitungsfunktion** $f'(x)$. Mit der h-Methode konnten wir die Differentialquotienten-Formel vereinfachen, sodass man anstatt zwei verschiedenen $x$-Werten nur noch ein $x$-Wert einsetzen muss. Somit konnten wir eine Ableitungsfunktion herleiten. Tatsächlich sind die in der Praxis verwendeten Ableitungsfunktionen noch viel bequemer zu nutzen als die mit der h-Methode erweiterte Differentialquotienten-Formel. Diese werden wir uns in den nächsten Kapiteln anschauen.
