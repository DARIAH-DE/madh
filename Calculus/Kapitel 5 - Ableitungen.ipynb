## 2. Ableitungsregeln <a class="anchor" id="2"/>

Die Berechnungen von Ableitungen sind für die folgenden Themen dieser Tutorialreihe essenziell. Dies bedeutet, dass Sie in vielen Aufgaben von Funktionen die Ableitungen bilden müssen. Mit der im vorherigen Kapitel eingeführten **h-Methode** ist die Berechnung der Ableitung einer Funktion zwar möglich, aber sehr aufwendig. Es gibt daher **Ableitungsregeln**, welche die Ableitungsberechnungen vereinfachen. Eine Herleitung dieser Regeln wird es in diesem Kapitel nicht geben, es werden lediglich die Regeln aufgestellt. Diese <u>müssen</u> bekannt sein, d.h. Sie müssen diese Regeln **auswendig** lernen bzw. anwenden können. Aber keine Sorge, diese Regeln sind nicht sonderlich schwer und behandeln Schulstoff. Im weiteren Verlauf wird es immer mehr Rechenregeln geben, die Sie auswendig kennen und vor allem auch anwenden können müssen. Diese Regeln werden in einem sogenannten **Werkzeugkasten** kurz zusammengefasst, auf welchen Sie bei Bedarf immer wieder zurückgreifen können. Dieser Werkzeugkasten erfüllt mehrere Zwecke: Zum einen wissen Sie so immer sofort, welche Regeln Sie im Gedächtnis behalten sollen. Damit Sie nicht immer zu den Notebooks mit den entsprechenden Werkzeugkästen springen müssen, liegt dieser Tutorialreihe ein **Cheatsheet** (dt.: Spickzettel) bei, bei der sie diese Regeln zusammengefasst finden können, um ihr Gedächtnis auffrischen zu können. Zum anderen ist der Zweck dieses Werkzeugkasten, dass Sie die in ihm enthaltenen Regeln nicht als lästige Belastung (AW TODO), sondern vielmehr als eine Reihe von mathematischen **Werkzeugen** ansehen. Diese Werkzeuge sollen Ihnen Rechenschritte erleichtern, so wie ein Hammer dabei hilft, einen Nagel in die Wand zu schlagen, ein Vorgang, welcher ohne den Hammer unpraktisch, ineffizient und oft auch unmöglich ist. 

In diesem Kapitel werden wir mehrere Werkzeugkästen kennen lernen. Bei beiden Werkzeugkästen werden nur die Regeln dargestellt, die Erläuterungen mit Beispielen folgen dann. 

<div class="tool">
<span class="invisible_border"><img src="img/gear.png" class="icon"/><b>Werkzeugkasten: <u>Ableitungsregeln I - Konstanten, x, Produktregel, Faktorregel</u></b></span>
    <br>
    <br>
1. <b>Ableitung einer Konstanten</b>: Die Ableitung einer Konstanten ist $0$.<br><br>
<div style="text-indent:20px;"> $f(x) = c$ &nbsp; &rarr; &nbsp; $f'(x) = 0$</div><br>    
2. <b>Ableitung von $x$</b>: Die Ableitung von $x$ ist $1$.<br><br>
<div style="text-indent:20px;"> $f(x) = x$ &nbsp; &rarr; &nbsp; $f'(x) = 1$</div><br>    
<div class="border">3. <b>Potenzregel</b>: Die Ableitung von $x^n$ ist $n \cdot x^{n-1}$.<br><br>
<div style="text-indent:20px;"> $f(x) = x^n$ &nbsp; &rarr; &nbsp; $f'(x) = n \cdot x^{n-1}$</div></div><br>
4. <b>Faktorregel</b>: Beim Ableiten bleibt der konstante Faktor unverändert erhalten.<br><br>
<div style="text-indent:20px;"> $f(x) = c \cdot g(x)$ &nbsp; &rarr; &nbsp; $f'(x) = c \cdot g'(x)$</div><br>
<div class="border">5. <b>Summenregel / Differenzregel</b>: Eine Summe/Differenz wird abgeleitet, indem man jeden Summanden/Minuend/Subtrahend für sich ableitet und die Ableitungen addiert. Beim Subtrahenden bleibt das negative Vorzeichen bestehen.<br><br>
<div style="text-indent:20px;"> $f(x) = g(x) + h(x)$ &nbsp; &rarr; &nbsp; $f'(x) = g'(x) + h'(x)$</div>
<div style="text-indent:20px;"> $f(x) = g(x) - h(x)$ &nbsp; &rarr; &nbsp; $f'(x) = g'(x) - h'(x)$</div></div><br>
